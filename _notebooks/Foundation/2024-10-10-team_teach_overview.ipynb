<h1>YAYAYAY SO FUN AP CSA Units 1-9 Summary</h1>

<details>
<summary><strong>Unit 1: Primitive Types</strong></summary>
<ul>
  <li><strong>Basic Data Types</strong>: int, double, boolean, char</li>
  <li><strong>Variable Declaration</strong>: Example: <code>int x = 5;</code></li>
  <li><strong>Arithmetic Operators</strong>: +, -, *, /, %, with precedence rules (PEMDAS)</li>
  <li><strong>Type Casting</strong>: Converting between <code>int</code> and <code>double</code> (note truncation for double to int)</li>
  <li><strong>Input Handling</strong>: <code>Scanner sc = new Scanner(System.in);</code></li>
  <li><strong>Output Handling</strong>: <code>System.out.println();</code></li>
</ul>
<p>Primitives = monkeys = simple</p>
<p>Basically: primitives are the most basic data types which are stored as their actual values in variables and not just a memory address.</p>
</details>

<hr>

<details>
<summary><strong>Unit 2: Working with Objects</strong></summary>
<ul>
  <li><strong>Object Creation</strong>: Objects instantiated with <code>new</code></li>
  <li><strong>Strings</strong>: Example: <code>String str = "hello";</code></li>
  <ul>
    <li><strong>Common Methods</strong>: <code>length()</code>, <code>substring()</code>, <code>indexOf()</code>, <code>charAt()</code></li>
  </ul>
  <li><strong>Null References</strong>: <code>null</code> as a placeholder for uninitialized objects</li>
  <li><strong>Equality Check</strong>: <code>==</code> for references, <code>.equals()</code> for content comparison</li>
  <li><strong>Math Operations</strong>: <code>Math.pow()</code>, <code>Math.sqrt()</code>, <code>Math.random()</code></li>
</ul>
<p>Objects are just like instantiations of a class; they create a new "instance" of the class that it's creating/calling to create. This means that methods that are defined within that class are within the variable the object is stored in.</p>
</details>

<hr>

<details>
<summary><strong>Unit 3: Boolean Expressions and If Statements</strong></summary>
<ul>
  <li><strong>Boolean Operators</strong>: <code>&&</code>, <code>||</code>, <code>!</code></li>
  <li><strong>Comparison Operators</strong>: <code>&lt;</code>, <code>&gt;</code>, <code>&lt;=</code>, <code>&gt;=</code>, <code>==</code>, <code>!=</code></li>
  <li><strong>Conditional Statements</strong>:</li>
  <ul>
    <li><code>if</code>, <code>else if</code>, <code>else</code></li>
    <li><strong>Nesting</strong>: Embedding conditions within other conditions</li>
  </ul>
  <li><strong>Logical Expressions</strong>: Combine conditions for complex decision-making</li>
</ul>
<p>Basically: and -&gt; <code>&&</code>, or -&gt; <code>||</code>, not -&gt; <code>!</code>, and if statements have parentheses and brackets ü§Ø</p>
</details>

<hr>

<details>
<summary><strong>Unit 4: Loops and Iteration</strong></summary>
<ul>
  <li><strong>Types of Loops</strong>:</li>
  <ul>
    <li><code>while(condition) { }</code></li>
    <li><code>for(initialization; condition; update) { }</code></li>
    <li><code>do { } while(condition);</code></li>
  </ul>
  <li><strong>Break/Continue</strong>: Interrupt or skip loop iterations</li>
  <li><strong>Nested Loops</strong>: Loops within loops</li>
  <li><strong>Use Cases</strong>: Array traversal, searching, counting operations</li>
</ul>
<p>For-each loops are a bit different:<br><code>for (type var : array) {}</code></p>
</details>

<hr>

<details>
<summary><strong>Unit 5: Creating Classes</strong></summary>
<ul>
  <li><strong>Class Structure</strong>: Fields, methods, constructors</li>
  <li><strong>Constructor</strong>: Used to initialize an object's state</li>
  <ul>
    <li>Example: <code>public ClassName(type parameter) { }</code></li>
  </ul>
  <li><strong>Getters</strong>: Access field values</li>
  <li><strong>Setters</strong>: Modify field values</li>
  <li><strong>Static Members</strong>: Shared across all instances of the class</li>
  <li><strong>Encapsulation</strong>: Protect fields using private access and public methods</li>
</ul>
<p>Getters and setters are just methods that.... wait for it.... GET AND SET</p>
<p>Constructor: <code>public</code>, <code>private</code>, <code>static</code>, <code>void</code> etc.</p>
</details>

<hr>

<details>
<summary><strong>Unit 6: Arrays</strong></summary>
<ul>
  <li><strong>Array Declaration</strong>: Example: <code>type[] arrayName = new type[size];</code></li>
  <li><strong>Accessing Elements</strong>: <code>array[index]</code></li>
  <li><strong>Traversing Arrays</strong>: Use loops to iterate through elements</li>
  <li><strong>Common Operations</strong>: Finding max, min, sum, and average</li>
  <li><strong>Bounds Checking</strong>: Prevent <code>ArrayIndexOutOfBoundsException</code></li>
</ul>
<p>Arrays: Don't ask for more than you have, else you get an <code>ArrayIndexOutOfBoundsException</code>. Unlike Collegeboard, Java is sane and starts indexing at 0.</p>
</details>

<hr>

<details>
<summary><strong>Unit 7: ArrayList</strong></summary>
<ul>
  <li><strong>Declaration</strong>: <code>ArrayList&lt;Type&gt; list = new ArrayList&lt;&gt;();</code></li>
  <li><strong>Key Methods</strong>:</li>
  <ul>
    <li><code>add()</code>, <code>remove()</code>, <code>set()</code>, <code>get()</code>, <code>size()</code></li>
  </ul>
  <li><strong>Dynamic Resizing</strong>: ArrayList can grow or shrink in size</li>
  <li><strong>Iteration</strong>: Use for-each or standard for loops</li>
  <li><strong>Autoboxing/Unboxing</strong>: Automatic conversion between primitives and wrapper types</li>
</ul>
<p>Benefits over <code>List</code>: You can dynamically modify, especially useful when modifying while iterating.</p>
</details>

<hr>

<details>
<summary><strong>Unit 8: 2D Arrays</strong></summary>
<ul>
  <li><strong>Declaration</strong>: <code>type[][] arrayName = new type[rows][cols];</code></li>
  <li><strong>Element Access</strong>: <code>array[row][col]</code></li>
  <li><strong>Traversing</strong>: Use nested loops for row-column iteration</li>
  <li><strong>Applications</strong>: Common in grids and matrix-related problems</li>
  <li><strong>Algorithms</strong>: Processing row-wise or column-wise</li>
</ul>
<p>2D Arrays: Basically chonkier arrays. An array within an array. THE MATRIX.</p>
<p><img src="https://github.com/user-attachments/assets/b74499e9-4a63-4bec-b079-3eda6a58e468" alt="Matrix image"></p>
</details>

<hr>

<details>
<summary><strong>Unit 9: Inheritance</strong></summary>
<ul>
  <li><strong>Inheritance Basics</strong>: <code>class Subclass extends Superclass</code></li>
  <li><strong>Subclass</strong>: Inherits fields and methods from the superclass</li>
  <li><strong>Overriding Methods</strong>: Redefining a method from the superclass in a subclass</li>
  <li><strong>Super Keyword</strong>: Used to call the superclass constructor or methods</li>
  <li><strong>Polymorphism</strong>: Using a superclass reference for a subclass object</li>
  <li><strong>Object Class</strong>: All classes derive from <code>Object</code></li>
</ul>
<p>Inheritance is kinda weird, but you can make your class and program structure like a tree and use the trunk to make leaves. By using <code>super</code>, you're essentially calling the superclass constructor and using its methods.</p>
</details>
